Exercício 01

```python
print("Digite dois numeros:")
x = input()
y = input()

while x != y:
    if x > y:
        print("Decrescente")
        print("Digite outros dois numeros:")
        x = input()
        y = input()
    if x < y:
        print("Crescente")
        print("Digite outros dois numeros:")
        x = input()
        y = input()
```

Exercicio 02

```python
soma = 0
count = 0
    
print("Digite as idades (digite um número negativo para finalizar):")
    
while True:
    idade = int(input())
    
    if count == 0 and idade < 0:
        print("Impossível calcular!")
    
    if idade < 0:
        break
        
    soma += idade
    count += 1

if count > 0:
    media = soma / count
    print(f"Média = {media:.2f}")
else:
    print("Impossível calcular")
```

Exercicio 04

```python
print("Digite os valores das coordenadas X e Y:")

x = int(input())
y = int(input())

while (x != 0 or y != 0):
    if (x > 0 and y > 0):
        print("1o Quadrante")
    elif (x < 0 and y > 0):
        print("2o Quadrante")
    elif (x < 0 and y < 0):
        print("3o Quadrante")
    elif (x > 0 and y < 0):
        print("4o Quadrante")
    
    print("Digite os valores das coordenadas X e Y:")

    x = int(input())
    y = int(input())

    if (x == 0 or y == 0):
        break
```

Exercicio 05

```python
nota1 = -1
nota2 = -1

while nota1 < 0 or nota1 > 10:
    nota1 = float(input())
    if nota1 < 0 or nota1 > 10:
        print("Valor inválido! Tente novamente: ")

while nota2 < 0 or nota2 > 10:
    nota2 = float(input())
    if nota2 < 0 or nota2 > 10:
        print("Valor inválido! Tente novamente: ")

media = (nota1 + nota2) / 2
print(f"media = {media:.2f}")
```

Exercicio 09

```python
x = int(input())
y = int(input())

a = min(x, y)
b = max(x, y)

soma = 0
atual = a + 1

while atual < b:
    if atual % 2 != 0:
        soma += atual
    atual += 1

print(soma)
```

Exercicio 12

```python
x = int(input("Quanto números você vai digitar?"))

for i in range(x):
    num = int(input("Digite um número: "))

    if (num == 0):
        print("Nulo")
    elif (num < 0 and num % 2 != 0):
        print("Ímpar negativo")
    elif (num < 0 and num % 2 == 0):
        print("Par negativo")
    elif (num > 0 and num % 2 != 0):
        print("Ímpar positivo")
    elif (num > 0 and num % 2 == 0):
        print("Par positivo")    
```

Exercicio 13

```python
x = int(input("Quanto casos você vai digitar?"))

for i in range(x):
    print("Digite três números: ")
    a = float(input())
    b = float(input())
    c = float(input())

    media = ((2 * a) + (3 * b) + (5 * c)) / 10
    print(f"Média: {media:.1f}")
```

Exercicio 14

```python
x = int(input("Quanto casos você vai digitar?"))

for i in range(x):
    num = int(input("Entre com o numerador: "))
    den = int(input("Entre com o denominador: "))

    if (den == 0):
        print("Divisão impossível!")
    elif (num == 0):
        print("Divisão = 0.00")
    elif (num == 0 and den == 0):
        print("Indeterminação matemática")
    else:
        div = num / den
        print(f"Divisão = {div:.2f}")
```

Exercicio 15

```python
n = -1
while n < 0 or n > 15:
    n = int(input("Digite um número de 0 a 15: "))
    if n < 0 or n > 15:
        print("Número inválido! Deve ser entre 0 e 15.")
        

fatorial = 1
for i in range(2, n + 1):
    fatorial *= i

print(f"{n}! = {fatorial}")
```

Exercicio 16

```python
x = int(input("Quantos casos de testes serão digitados? "))
qtdeC = 0
qtdeR = 0
qtdeS = 0
qtdeT = 0

for i in range(x):
    qtde = int(input("Quantidade de cobaias? "))
    tipo = input("Tipo de cobaia (R:Rato, S:Sapo, C:Coelho): ").strip().upper()

    if tipo == 'C':
        qtdeC += qtde
    elif tipo == 'R':
        qtdeR += qtde
    elif tipo == 'S':
        qtdeS += qtde

    qtdeT += qtde

print("\nRELATÓRIO FINAL:")
print(f"Total: {qtdeT} cobaias")
print(f"Total de coelhos: {qtdeC}")
print(f"Total de ratos: {qtdeR}")
print(f"Total de sapos: {qtdeS}")
print(f"Percentual de coelhos: {(qtdeC / qtdeT * 100):.2f}%")
print(f"Percentual de ratos: {(qtdeR / qtdeT * 100):.2f}%")
print(f"Percentual de sapos: {(qtdeS / qtdeT * 100):.2f}%")
```
