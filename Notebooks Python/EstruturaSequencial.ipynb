## Exercícios 01 - Estrutura Sequencial

Exercício 01

```python
largura = float(input("Digite a largura do terreno: "))
comprimento = float(input("Digite o comprimento do terreno: "))
valMtQuad = float(input("Digite o valor do metro quadrado: "))

area = largura * comprimento
valFinal = area * valMtQuad

print(f"Area do terreno: {area:.2f}\r\nPreco do terreno: {valFinal:.2f}")
```

Exercício 02

```python
import math as m

base = float(input("Informe a base do retângulo: "))
altura = float(input("Informe a altura do retângulo: "))

area = base * altura
perimetro = (base * 2) + (altura * 2)
diagonal = m.sqrt(m.pow(base, 2) + m.pow(altura, 2))

print(f"Area: {area:.4f}\r\nPerimetro: {perimetro:.4f}\r\nDiagonal: {diagonal:.4f}")
```

Exercício 03

```python
print("Dados da primeira pessoa:")
nomePessoa1 = input("Nome: ")
idadePessoa1 = int(input("Idade: "))

print("Dados da segunda pessoa:")
nomePessoa2 = input("Nome: ")
idadePessoa2 = int(input("Idade: "))

media = (idadePessoa1 + idadePessoa2) / 2

print(f"A idade média de {nomePessoa1} e {nomePessoa2} é {media:.1f}")
```

Exercício 05

```python
valUnit = float(input("Preço unitário do produto: "))
qtde = int(input("Quantidade comprada: "))
valReceb = float(input("Dinheiro recebido: "))

valTotal = qtde * valUnit

if (valReceb > valTotal):
    print(f"Troco: R$ {(valReceb - valTotal):.2f}")
elif (valReceb == valTotal):
    print("Troco: R$ 0.00")
elif (valReceb < valTotal):
    print(f"Está faltando R$ {((valReceb - valTotal) * (-1)):.2f} para completar o pagamento")
```

Exercício 06

```python
import math as m

r = float(input("Digite o valor do raio do círculo: "))
area = m.pi * m.pow(r, 2)

print(f"Área do círculo: {area:.3f}")
```

Exercício 09

```python
a = float(input("Digite a medida A: "))
b = float(input("Digite a medida B: "))
c = float(input("Digite a medida C: "))

areaQuad = a * a
areaTrig = (a * b) / 2
areaTrap = ((a + b) * c) / 2

print(f"Área do quadrado: {areaQuad:.4f}")
print(f"Área do triângulo retângulo: {areaTrig:.4f}")
print(f"Área do trapézio: {areaTrap:.4f}")
```
