Exercício 01

```python
texto = input("Frase: ").lower()
limpo = ''.join(c for c in texto if c.isalpha())

presentes = set(limpo)
alfabeto = set('abcdefghijklmnopqrstuvwxyz')

if presentes >= alfabeto:
    print("Esta frase é um pangrama!")
else:
    print("Esta frase não é pangrama. Faltam letras.")
```

Exercício 02

```python
palavra = input("Digite uma palavra: ").lower()
base = "python"

if len(palavra) != len(base):
    print("Não é um anagrama.")
else:
    isAnagrama = True

    for letra in base:
        if letra not in palavra:
            isAnagrama = False
            break

    if isAnagrama:
        print("É um anagrama de Python.")
    else:
        print("Não é um anagrama.")
```

Exercício 03

```python
senha = input("Digite uma senha: ")

if len(senha) < 8:
    print("Senha fraca: deve conter pelo menos 8 caracteres.")
elif not any(c.isupper() for c in senha):
    print("Senha fraca: deve conter pelo menos uma letra maiúscula.")
elif not any(c.islower() for c in senha):
    print("Senha fraca: deve conter pelo menos uma letra minúscula.")
elif not any(c.isdigit() for c in senha):
    print("Senha fraca: deve conter pelo menos um número.")
elif not any(c in "!@#$%^&*()-_=+[]{}|;:'\"<>,.?/" for c in senha):
    print("Senha fraca: deve conter pelo menos um caractere especial.")
else:
    print("Senha forte!")
```

Exercício 05

```python
numero = input("Digite um número de telefone válido (ex: (XX) XXXXX-XXXX):")
valido = False

if len(numero) == 14 or len(numero) == 15:
    if (
        numero[0] == "(" and
        numero[1:3].isdigit() and
        numero[3] == ")" and
        numero[4] == " " and
        numero[-5] == "-" and
        numero[-4:].isdigit()
    ):
        parte_principal = numero[5:-5]

        if (
            parte_principal.isdigit() and
            (len(parte_principal) == 4 or len(parte_principal) == 5)
        ):
            valido = True

print("Telefone válido?" , "Sim" if valido == True else "Não")
```

Exercício 07

```python
frase = input("Insira uma frase que seja um palíndromo: ")
frase = frase.lower()

acentos = "áàâãäéèêëíìîïóòôõöúùûüç"
sem_acentos = "aaaaaeeeeiiiiooooouuuuc"
nova = ""

for c in frase:
    if c in acentos:
        indice = acentos.index(c)
        nova += sem_acentos[indice]
    else:
        nova += c

limpa = ""

for c in nova:
    if c.isalnum():
        limpa += c

isPalindromo = limpa == limpa[::-1]

print("É um palíndromo?", "Sim" if isPalindromo == True else "Não")
```

Exercício 08

```python
cpf = input("Entre com o número de CPF (Apenas números): ")
valido = True

# Deve ter 11 digitos e ser numerico
if len(cpf) != 11 or not cpf.isdigit():
    valido = False

# Evita CPF com todos os números iguais
elif cpf == cpf[0] * 11:
    valido = False

else:
    soma = 0
    for i in range(9):
        soma += int(cpf[i]) * (10 - i)

    resto = (soma * 10) % 11
    if resto == 10:
        resto = 0

    # Se resto for diferente do 10º digito do cpf, o cpf é invalido
    # Pelo que entendi, isso é uma regra para garantir se o cpf é válido ou não...
    if resto != int(cpf[9]):
        valido = False

    soma = 0
    for i in range(10):
        soma += int(cpf[i]) * (11 - i)

    resto = (soma * 10) % 11
    if resto == 10:
        resto = 0

    if resto != int(cpf[10]):
        valido = False

print("CPF", "válido" if valido == True else "inválido")
```

Exercício 09

```python
texto = "abaaba"
contador = 0
n = len(texto)

for i in range(n):
    for j in range(i + 3, n + 1):  # tamanho mínimo 3
        substring = texto[i:j]
        
        if substring == substring[::-1]:
            contador += 1

print("Quantidade de substrings palíndromas (maior que 3):", contador)
```
