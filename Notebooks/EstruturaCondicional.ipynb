Exercício 02

```python
import math as m

a = float(input("Coeficiente a: "))
b = float(input("Coeficiente b: "))
c = float(input("Coeficiente c: "))

delta = m.pow(b, 2) - 4 * a * c

if (delta > 0):
    x1 = (-b + m.sqrt(delta)) / 2
    x2 = (-b - m.sqrt(delta)) / 2
    print(f"X1: {x1:.4f}")
    print(f"X2: {x2:.4f}")
elif (delta == 0):
    x1 = (-b + m.sqrt(delta)) / 2
    print(f"X1 = X2 = X: {x1:.4f}")
else:
    print("Esta equação não possui raízes reais!")
```

Exercício 05

```python
valUnit = float(input("Preço unitário do produto: "))
qtde = int(input("Quantidade comprada: "))
valReceb = float(input("Dinheiro recebido: "))

valTotal = qtde * valUnit

if (valReceb > valTotal):
    print(f"Troco: R$ {(valReceb - valTotal):.2f}")
elif (valReceb == valTotal):
    print("Troco: R$ 0.00")
elif (valReceb < valTotal):
    print(f"Valor insuficiente. Faltam R$ {((valReceb - valTotal) * (-1)):.2f}")
```

Exercício 08

```python
pgt = input("Voce vai digitar a temperatura em qual escala (C/F)?")

if (pgt == 'C'):
    tC = float(input("Digite a temperatura em Celsius: "))
    tF = 1.8 * tC + 32
    print(f"Temperatura equivalente em Fahrenheit: {tF:.2f}")
if (pgt == 'F'):
    tF = float(input("Digite a temperatura em Fahrenheit: "))
    tC = (5 * tF - 160) / 9
    print(f"Temperatura equivalente em Celsius: {tC:.2f}")
```

Exercicio 12

```python
hora_inicial = int(input("Insira a hora inicial (0-23): "))
minuto_inicial = int(input("Insira o minuto inicial (0-59): "))

hora_final = int(input("Insira a hora final (0-23): "))
minuto_final = int(input("Insira o minuto final (0-59): "))

inicio_em_minutos = hora_inicial * 60 + minuto_inicial
fim_em_minutos = hora_final * 60 + minuto_final

if fim_em_minutos > inicio_em_minutos:
    duracao_minutos = fim_em_minutos - inicio_em_minutos
else:
    duracao_minutos = (24 * 60 - inicio_em_minutos) + fim_em_minutos

duracao_horas = duracao_minutos // 60
duracao_min = duracao_minutos % 60

print(f"O jogo durou {duracao_horas} hora(s) e {duracao_min} minuto(s)")
```
